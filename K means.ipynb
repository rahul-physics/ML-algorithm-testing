```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('income.csv')
```

```python
df
```

```python
df=df.drop('Name',axis='columns')
df
```

```python
from sklearn.cluster import KMeans
kmean=KMeans(n_clusters=3)
```

```python
y_predict=kmean.fit_predict(df)
```

```python
y_predict
```

```python
df['cluster']=y_predict
df
```

```python
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]
```

```python
plt.scatter(df1['Age'],df1['Income($)'],color='Red')
plt.scatter(df2['Age'],df2['Income($)'],color='Blue')
plt.scatter(df3['Age'],df3['Income($)'],color='Green')

```

```python
from sklearn.preprocessing import MinMaxScaler
mn=MinMaxScaler()
```

```python
mn.fit(df[['Age']])
df.Age=mn.transform(df[['Age']])
mn.fit(df[['Income($)']])
df[['Income($)']]=mn.transform(df[['Income($)']])
```

```python
df
```

```python
df=df.drop('cluster',axis='columns')
df
```

```python
y_predict=kmean.fit_predict(df)
```

```python
y_predict
```

```python
df['cluster']=y_predict
df
```

```python
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]
```

```python
plt.scatter(df1['Age'],df1['Income($)'],color='Red')
plt.scatter(df2['Age'],df2['Income($)'],color='Blue')
plt.scatter(df3['Age'],df3['Income($)'],color='Green')
plt.xlabel('Normlaized Age')
plt.ylabel('Normlaized salary')
```

```python
k_range=range(1,10)
sse=[]
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(df)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_range,sse)
```
