```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.scatter(df.area,df.price)
```

```python
plt.show()
```

```python
model=linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
d=pd.read_csv('areas.csv')
d
```

```python
model.predict(d)
```

```python
p=model.predict(d)
```

```python
d['predict']=p
```

```python
d
```

```python
d.to_csv('prediction.csv',index=False)
```

```python
df1=pd.read_csv('canada_per_capita_income.csv')
df1.head()
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df1[['year']],df1['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.scatter(df1['year'],df1['per capita income (US$)'])
```

```python
plt.show()
```

```python
plt.scatter(df1['year'],df1['per capita income (US$)'])
plt.plot(df1.year,reg.predict(df1[['year']]))
plt.show()
```

```python
df2=pd.read_csv('homepricesm.csv')
df2
```

```python
df2.bedrooms.median()
```

```python
df2.bedrooms=df2.bedrooms.fillna(df2.bedrooms.median())
```

```python
df2
```

```python
model1=linear_model.LinearRegression()
model1.fit(df2[['area','bedrooms','age']],df2.price)
```

```python
model1.predict([[3000,3,40]])
```

```python
import joblib
```

```python
joblib.dump(model1,'model1_joblib')
```

```python
m1=joblib.load('model1_joblib')
```

```python
m1.predict([[3000,3,40]])
```
