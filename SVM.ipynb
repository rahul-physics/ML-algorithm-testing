```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('titanic.csv')
data.head()
```

```python
X=data[['Pclass','Sex','Age','Fare']]
y=data['Survived']
X
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
X['Sex_encoded']=le.fit_transform(X['Sex'])
X
```

```python
X
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
df1=pd.DataFrame(iris.data,columns=iris.feature_names)
df1
```

```python
df1['target']=iris.target
df1
```

```python
df1[df1.target==0]
```

```python
Xi=df1.drop('target',axis='columns')
Xi
```

```python
yi=df1.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split=train_test_split(Xi,yi,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
model1=SVC()
```

```python
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```
